```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
from tpot import TPOTClassifier
```

```python
#Data organize
data = load_iris()
x = data.data
y = data.target
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
#Tpot Classifier Hyperparameters and Model define
tpot = TPOTClassifier(
    generations=10,
    population_size=100,
    offspring_size=100,
    mutation_rate=0.9,
    crossover_rate=0.1,
    scoring='accuracy',
    max_time_mins=2,
    early_stop=False,
    random_state=0,
    verbosity=2)
tpot.fit(X_train, y_train)
print('Melhor modelo:', tpot.fitted_pipeline_)
print('Acurácia:', tpot.score(X_test, y_test))
```

```python
#Evaluating all classifiers
tpot.evaluated_individuals_
```

```python
#Accuracy
predict_tpot = tpot.predict(X_test)
accuracy_score(y_test, predict_tpot)
```
