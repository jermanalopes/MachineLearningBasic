```python
import h2o
from h2o.automl import H2OAutoML
```

```python
h2o.init()
```

```python
#Data organize
data = h2o.import_file('insurance.csv')
data = data.drop("C1")
trein, test = data.split_frame(ratios=[0.7], seed=1234)
```

```python
#AutoML Model Definition
model_autoML = H2OAutoML(max_runtime_secs=180, sort_metric='auto')
model_autoML.train(y='Accident', training_frame=trein)
```

```python
#Listing the best models
ranking = model_autoML.leaderboard
ranking.head()
```

```python
#Predict Classes with the best model
predict_h2o = model_autoML.leader.predict(test)
predict_h2o.head()
```
